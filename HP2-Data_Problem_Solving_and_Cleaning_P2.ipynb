## 4. STEP 3 - Creating of a function

Now we are going to package all the process inside of a reusable function.

Instructions:

Create a function called clean_user(user) that realize this actions:

1. Clean spaces and underscores from the names and divide them in name/lastname.

2. Convert the age to integer.

3. Convert the categories to lowercase.

Use the same code that you write in the last step. I just need to adjust itto work inside of a function.

```python
def clean_patients(patients):
    patients_clean = []
    for u in patients:
        name_clean = u[1].strip().replace("_", " ").split()
        age_clean = int(u[2])
        lowered_service = [service.lower() for service in u[3]]
        expenses_clean = [int(x) for x in u[4]]
        patient_clean = [int(u[0]), name_clean, age_clean, lowered_service, expenses_clean]
        patients_clean.append(patient_clean)
    return patients_clean

patients_clean = clean_patients(patients)
for patient in patients_clean:
    print(patient)#I tested it, so I can make sure it works
```

```python
dataframe_patients_clean = pd.DataFrame(patients_clean, columns=["ID", "Name", "Age", "Services", "Expenses"])
print(dataframe_patients_clean)
```

## 9. STEP 8 - Function to filter by service.

To automate the analysis by category implement a reusable function that allows to obtain the patients that had used an specific service.

Instructions:

1. Complete the function get_patients_by_category with two arguments:
       - A list of patients and a category to search for.

2. Inside the function:
        - Create an empty list to save the results.
        - Go throught every patient with a loop for.
        - Use an if condition to verify if the category it's in the services list.
        -If  the condition is achieved, add to the list of results to a sublist with the
         ID, complete name, age and total expence.

3. Return the resulting list at the end of the function.

```python
def patients_by_service(patients, service):
    result = []
    for p in patients:
        patient_id = p[0]
        patient_name = p[1]
        patient_age = p[2]
        services = p[3]
        expenses = p[4]

        if service in services:
            result.append([patient_id, patient_name, patient_age, sum(expenses)])
    return result

patients_service = patients_by_service(patients_clean, 'emergency')
for patient in patients_service:
    print(patient)
```
